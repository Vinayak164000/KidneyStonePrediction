```python
def remove_outlier(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3-Q1
    lower_limit = Q1 - 1.5*IQR
    upper_limit = Q3 + 1.5*IQR
    return upper_limit, lower_limit
```

```python
cols = ["gravity","ph","osmo","cond","urea","calc"]
for col in cols:
    remove_outlier(train[col])
    if(train[col]>lower_limit and train[col]<upper_limit):
        train.drop(ph, axis = 1)
```

```python
X = train.drop('target', axis = 1)
Y = train['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression(random_state = 0)
regressor.fit(X_train,Y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_pred, Y_test)
print(cm)
accuracy_score(y_pred, Y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5,metric = 'minkowski', p = 2)
classifier.fit(X_train,Y_train)
```

```python
y_pred1 = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_pred1, Y_test)
print(cm)
accuracy_score(y_pred1, Y_test)
```

```python
from sklearn.svm import SVC
classifier2 = SVC(kernel = 'linear', random_state =0)
classifier2.fit(X_train,Y_train)
```

```python
y_pred2 = classifier2.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_pred2, Y_test)
print(cm)
accuracy_score(y_pred2, Y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier3 = GaussianNB()
classifier3.fit(X_train,Y_train)
```

```python
y_pred3 = classifier3.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_pred3, Y_test)
print(cm)
accuracy_score(y_pred3, Y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier4 = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier4.fit(X_train, Y_train)
```

```python
y_pred4 = classifier4.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_pred4, Y_test)
print(cm)
accuracy_score(y_pred4, Y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier5 = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier5.fit(X_train, Y_train)
```

```python
y_pred5 = classifier5.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_pred5, Y_test)
print(cm)
accuracy_score(y_pred5, Y_test)
```
